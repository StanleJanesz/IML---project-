```python
import matplotlib.pyplot as plt
```

```python
def plot_loss(path, stride, iters):
    iterations = []
    loss_values = []
    i = 0
    j = 0
    # Open the file and read line by line
    with open(file_path, 'r') as file:
        for line in file:
            if "Finished Training" in line:
                j=j+1
            if "loss:" in line and i%stride==0:  # Only process lines containing 'loss:'
                parts = line.split('loss: ')
                # Extract the iteration number (last value inside square brackets)
                iteration = int(parts[0].split(',')[1].strip('] ')) + (iters * (int(parts[0].split(',')[0].strip('[')) - 1 + j*10))
                # Extract the loss value (after 'loss: ')
                loss = float(parts[1])
                iterations.append(iteration)
                loss_values.append(loss)
            i=i+1

    # Plot the data
    plt.figure(figsize=(8, 5))
    plt.plot(iterations, loss_values, marker='o', linestyle='-', color='b')
    plt.xlabel('Iterations')
    plt.ylabel('Loss')
    plt.title('Loss over Iterations')
    plt.grid(True)
    plt.show()
```
